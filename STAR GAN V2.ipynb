# StarGAN V2

StarGAN v2 는 하나의 모델로 여러 도메인의 이미지를 생성할 수 있는 GAN 모델이다. 기존의 모델은 변경하고자 하는 도메인 개수만큼의 모델이 필요했던 반면, V2 는 image to image translation 이 가능한, 다양한 도메인에 대한 style transfer 가 가능하다.

#### 따라서 다양한 이미지를 생성하며, 다양한 도메인으로 확장할 수 있다는 점, 그리고 성능이 더 우수하다는 점에서 기존의 GAN 모델보다 발전했다고 할 수 있다.
